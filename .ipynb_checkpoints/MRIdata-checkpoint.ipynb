```python
from mrcnn.utils import Dataset

class TumorDataset(Dataset):
    def load_tumor(self, dataset_dir, subset):
        """Load a subset of the Brain Tumor dataset."""
        self.add_class("tumor", 1, "tumor")

        assert subset in ["TRAIN", "VAL"]
        dataset_dir = os.path.join(dataset_dir, subset)

        image_ids = next(os.walk(dataset_dir))[2]
        for image_id in image_ids:
            self.add_image(
                "tumor",
                image_id=image_id,
                path=os.path.join(dataset_dir, image_id)
            )

    def load_mask(self, image_id):
        """Load instance masks for an image."""
        info = self.image_info[image_id]
        mask_path = info["path"].replace(".jpg", "_mask.png")
        
        mask = skimage.io.imread(mask_path, as_gray=True)
        mask = np.where(mask > 0, 1, 0)  # Convert to binary mask
        mask = np.expand_dims(mask, axis=-1)  # Add an extra dimension
        
        return mask.astype(np.bool_), np.array([1], dtype=np.int32)

print("‚úÖ Custom dataset class defined successfully!")
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Image dimensions
img_size = (256, 256)  # Ensure consistency

# Function to create a mask from polygon coordinates
def create_mask_from_annotation(annotation, img_size):
    mask = np.zeros(img_size, dtype=np.uint8)  # Create a black mask

    for region in annotation["regions"]:
        shape_attr = region["shape_attributes"]
        
        if shape_attr["name"] == "polygon":
            all_x = np.array(shape_attr["all_points_x"])
            all_y = np.array(shape_attr["all_points_y"])

            # Scale the coordinates if needed (assuming original annotation is for a larger image)
            all_x = (all_x * img_size[1] / 256).astype(int)  # Adjust based on dataset size
            all_y = (all_y * img_size[0] / 256).astype(int)

            # Convert to contour format for OpenCV
            contour = np.array(list(zip(all_x, all_y)), dtype=np.int32)

            # Draw white tumor mask on black background
            cv2.fillPoly(mask, [contour], 1)

    return mask

# Load the original image
sample_image_path = r"C:\Users\Suraj Yadav\FinalProjectTY\datasetMRI\Br35H-Mask-RCNN\TRAIN\y0.jpg"  # Change path if needed
sample_image = load_img(sample_image_path, target_size=img_size, color_mode="grayscale")
sample_image = img_to_array(sample_image) / 255.0  # Normalize

# Generate mask for the sample annotation
sample_mask = create_mask_from_annotation(sample_annotation, img_size)

# Plot both images side by side
plt.figure(figsize=(10, 5))

# Original MRI scan
plt.subplot(1, 2, 1)
plt.imshow(sample_image.squeeze(), cmap="gray")
plt.title("Original MRI Scan")
plt.axis("off")

# Generated Tumor Mask
plt.subplot(1, 2, 2)
plt.imshow(sample_mask, cmap="gray")
plt.title("Generated Tumor Mask")
plt.axis("off")

plt.show()
```

```python
import os
import numpy as np
import cv2
from tensorflow.keras.preprocessing.image import load_img, img_to_array, save_img

# Define dataset directories
base_folder = r"C:\Users\Suraj Yadav\FinalProjectTY\datasetMRI\Br35H-Mask-RCNN"
processed_folder = os.path.join(base_folder, "Processed_Images")
image_folder = os.path.join(processed_folder, "images")
mask_folder = os.path.join(processed_folder, "masks")

# Create folders if they don't exist
os.makedirs(image_folder, exist_ok=True)
os.makedirs(mask_folder, exist_ok=True)

# Load the annotation JSON
import json
annotation_file = os.path.join(base_folder, "annotations_all.json")

with open(annotation_file, "r") as f:
    annotations = json.load(f)

# Image size
img_size = (256, 256)

# Function to create mask from annotation
def create_mask_from_annotation(annotation, img_size):
    mask = np.zeros(img_size, dtype=np.uint8)  # Black background

    for region in annotation["regions"]:
        shape_attr = region["shape_attributes"]
        
        if shape_attr["name"] == "polygon":
            all_x = np.array(shape_attr["all_points_x"])
            all_y = np.array(shape_attr["all_points_y"])

            # Scale the coordinates to match resized image
            all_x = (all_x * img_size[1] / 256).astype(int)  # Adjust if needed
            all_y = (all_y * img_size[0] / 256).astype(int)

            # Create contour
            contour = np.array(list(zip(all_x, all_y)), dtype=np.int32)

            # Draw the mask (White Tumor)
            cv2.fillPoly(mask, [contour], 1)

    return mask

# Process and save images/masks
for key, annotation in annotations.items():
    filename = annotation["filename"]

    # Define file paths
    original_image_path = os.path.join(base_folder, "TRAIN", filename)  # Change TRAIN if needed
    processed_image_path = os.path.join(image_folder, filename)
    processed_mask_path = os.path.join(mask_folder, filename)

    # Check if file exists
    if os.path.exists(original_image_path):
        # Load and normalize the image
        img = load_img(original_image_path, target_size=img_size, color_mode="grayscale")
        img_array = img_to_array(img) / 255.0  # Normalize
        
        # Generate the mask
        mask = create_mask_from_annotation(annotation, img_size)

        # ‚úÖ Expand mask dimensions to make it (150, 150, 1)
        mask = np.expand_dims(mask, axis=-1)

        # Save processed images & masks
        save_img(processed_image_path, img_array)
        save_img(processed_mask_path, mask)

print(f"‚úÖ All images processed and saved in: {image_folder}")
print(f"‚úÖ All masks processed and saved in: {mask_folder}")
```

```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.model_selection import train_test_split

# Define paths to processed image and mask folders
image_folder = r"C:\Users\Suraj Yadav\FinalProjectTY\datasetMRI\Br35H-Mask-RCNN\Processed_Images\images"
mask_folder = r"C:\Users\Suraj Yadav\FinalProjectTY\datasetMRI\Br35H-Mask-RCNN\Processed_Images\masks"

# Image size
img_size = (256, 256)  # Keep consistent with preprocessing

# Load images and masks
data = []
labels = []

image_files = sorted(os.listdir(image_folder))
mask_files = sorted(os.listdir(mask_folder))

for img_name, mask_name in zip(image_files, mask_files):
    img_path = os.path.join(image_folder, img_name)
    mask_path = os.path.join(mask_folder, mask_name)

    # Load and normalize images
    img = load_img(img_path, target_size=img_size, color_mode="grayscale")
    img = img_to_array(img) / 255.0  # Normalize

    # Load and normalize masks (ensure binary format 0 and 1)
    mask = load_img(mask_path, target_size=img_size, color_mode="grayscale")
    mask = img_to_array(mask)
    mask = np.where(mask > 0, 1, 0)  # Convert to binary mask (0 for background, 1 for tumor)

    data.append(img)
    labels.append(mask)

# Convert to NumPy arrays
data = np.array(data, dtype=np.float32)
labels = np.array(labels, dtype=np.float32)

# Split data into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

print(f"‚úÖ Training data shape: {X_train.shape}")
print(f"‚úÖ Testing data shape: {X_test.shape}")
```

```python
from mrcnn.config import Config

class TumorConfig(Config):
    NAME = "tumor_segmentation"

    # Number of classes (1 background + 1 tumor)
    NUM_CLASSES = 1 + 1  

    # GPU settings
    GPU_COUNT = 1
    IMAGES_PER_GPU = 2  # Adjust based on available memory

    # Training parameters
    STEPS_PER_EPOCH = 100  # Adjust based on dataset size
    VALIDATION_STEPS = 10

    # Image properties
    IMAGE_MIN_DIM = 256
    IMAGE_MAX_DIM = 256

    # Minimum confidence for detection
    DETECTION_MIN_CONFIDENCE = 0.8

# Initialize configuration
config = TumorConfig()
config.display()
```
